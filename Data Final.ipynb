```python
def pie(category):
    """This function generates a pie chart showing the distribution of customer behavior based on the given category.

    Args:
       category: str. The column name representing the types of education in the customer behavior dataset.

    Returns:
       None
    """
    labels = []
    label_percentages = []

    for cat in cd[category].unique():
        labels.append(cat)
        behavior_count = cd[cd[category] == cat].shape[0]
        label_percentages.append(behavior_count)


    fig, axs = plt.subplots()
    axs.pie(label_percentages, labels=labels, autopct='%1.1f%%')
    plt.title(category)
    plt.show()
```
